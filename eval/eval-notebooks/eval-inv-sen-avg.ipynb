```python
import matplotlib
%matplotlib inline
import numpy as np
import time

import sys
sys.path.append('../../core')
sys.path.append('../../configuration-setup/')

from evaluation import Evaluation

evalObject = Evaluation(dynamics='Brussellator')
dataObject = evalObject.setDataObject()
evalObject.setIterCount(5)
evalObject.setNStates(100)

time_taken = 0.0
for idx in range(10):
    start = time.time()
    ref_traj = dataObject.generateTrajectories(samples=1)[0]
    time_step = 51

    dest = ref_traj[time_step]
    dests = [dest]
    evalObject.reachDestInv(dests=dests, d_time_steps=[time_step], threshold=0.001)
    time_taken += time.time() - start
print("time taken {}".format(time_taken/10))
```
