```python
import matplotlib
%matplotlib inline

from learningModule import DataConfiguration, CreateTrainNN

dataObject = DataConfiguration(dynamics='Brussellator')
dataObject.setSteps(500)
dataObject.setSamples(30)
dataObject.setLowerBound([0.0, 0.0]) 
dataObject.setUpperBound([10.0, 10.0])
dataObject.generateTrajectories()
dataObject.showTrajectories()
dataObject.createData(jumps=[1,2,4,8,10,12,16])

nnObject = CreateTrainNN(dynamics='Brussellator')
nnObject.createInputOutput(data_object = dataObject, inp_vars=['x','xp','v','t'], out_vars=['vp'])
nnObject.setEpochs(30)
nnObject.trainTestNN(optim='SDG', loss_fn='mae', layers=8, neurons=512)
```
