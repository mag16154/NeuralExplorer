```python
import matplotlib
%matplotlib inline
import numpy as np
import time

import sys
sys.path.append('../../core')
sys.path.append('../../configuration-setup/')

from evaluation import Evaluation

evalObject = Evaluation(dynamics='MC')
dataObject = evalObject.setDataObject()
dataObject.setLowerBound([-0.60, 0.0])
dataObject.setUpperBound([-0.40, 0.0])
evalObject.setIterCount(5)
evalObject.setNStates(10)

ref_traj = dataObject.generateTrajectories(samples=1)[0]
time_step = 40

time_taken = 0.0
dest = ref_traj[time_step]
dests=[]
start = time.time()
dests.append([-0.8, 0.0])
dests.append([-0.9, 0.0])
dests.append([-1.0, 0.0])
dests.append([-1.1, 0.0])
dests.append([-1.2, 0.0])
dests.append([-1.3, 0.0])
dests.append([-1.4, 0.0])
evalObject.reachDestInv(dests=dests, d_time_steps=[time_step-15,time_step+25], threshold=0.01)
time_taken = time.time() - start
print("time taken {}".format(time_taken))
```
