```python
import matplotlib
%matplotlib inline
import numpy as np


from evaluation import Evaluation

evalObject = Evaluation(dynamics='Brussellator')
dataObject = evalObject.setDataObject()
evalObject.setIterCount(3)
evalObject.setNStates(1 )

ref_traj = dataObject.generateTrajectories(samples=1)
ref_traj = ref_traj[0]
time_step = 49
#print(ref_traj[0], ref_traj[time_step])

dest = ref_traj[time_step]
dests = [dest]
#dests.append([9.40636489, 0.15787557])
evalObject.reachDest(dests=dests, d_time_step=time_step, threshold=0.1)
```
